```python
import os
from pathlib import Path
import torch
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm
import json
import zipfile
```

```python


# Путь к ZIP-файлу на диске
zip_path = "/content/drive/MyDrive/dataset.zip"
extract_to = "/content/dataset"

# Распаковка
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print("ZIP успешно распакован в:", extract_to)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
# Загружаем быструю модель
model = SentenceTransformer("cointegrated/rubert-tiny2")

# Путь к отзывам
base_dir = Path("/content/dataset/dataset")  # уточни путь при необходимости
label_map = {"neg": 0, "neu": 1, "pos": 2}

# Сбор всех текстов
results = []
for label, label_id in label_map.items():
    folder = base_dir / label
    for file_path in tqdm(list(folder.glob("*.txt")), desc=f"Обработка {label}"):
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read().strip()
        embedding = model.encode(text).tolist()
        results.append({
            "filename": file_path.name,
            "label": label_id,
            "embedding": embedding
        })

# Сохраняем
with open("/content/fast_russian_embeddings.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print(" Быстрые эмбеддинги сохранены.")
```
